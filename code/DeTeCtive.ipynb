```python
import json

def patch_src(input_path, output_path):
    with open(input_path, "r") as f:
        data = json.load(f)

    for row in data:
        label = str(row["label"])
        row["src"] = "gpt-3.5-turbo" if label == "1" else "human"

    with open(output_path, "w") as f:
        json.dump(data, f, indent=2)

# Run this
patch_src("/content/drive/MyDrive/fp/DeTeCtive/test_mini.json", "/content/drive/MyDrive/fp/DeTeCtive/test_mini.json")
patch_src("/content/drive/MyDrive/fp/DeTeCtive/train_mini.json", "/content/drive/MyDrive/fp/DeTeCtive/train_mini.json")
```

```python
import json

def check_bad_src(path, model_names):
    with open(path, "r") as f:
        data = json.load(f)

    for i, row in enumerate(data):
        if row["src"] not in model_names:
            print(f"Row {i} has bad src: {row['src']}")

model_name_dict = [
    "gpt-3.5-turbo", "text-davinci-002", "text-davinci-003",
    "13B", "30B", "65B", "7B", "GLM130B", "flan_t5_base", "flan_t5_large",
    "flan_t5_small", "flan_t5_xl", "flan_t5_xxl", "opt_1.3b", "opt_125m",
    "opt_13b", "opt_2.7b", "opt_30b", "opt_350m", "opt_6.7b", "opt_iml_30b",
    "opt_iml_max_1.3b", "bloom_7b", "t0_11b", "t0_3b", "gpt_j", "gpt_neox", "human"
]

check_bad_src("/content/drive/MyDrive/fp/DeTeCtive/test_mini.json", model_name_dict)
check_bad_src("/content/drive/MyDrive/fp/DeTeCtive/train_mini.json", model_name_dict)
```

```python
import json

def find_missing_or_null_src(file_path):
    with open(file_path, "r") as f:
        data = json.load(f)

    for i, row in enumerate(data):
        if "src" not in row or row["src"] is None:
            print(f"❗ Row {i} is missing or has null src → {row}")

find_missing_or_null_src("/content/drive/MyDrive/fp/DeTeCtive/test_mini.json")
find_missing_or_null_src("/content/drive/MyDrive/fp/DeTeCtive/train_mini.json")
```

```python
import json
from sklearn.model_selection import train_test_split

def clean_and_split(train_path, test_path):
    # Load raw data
    with open(train_path, 'r') as f:
        train_raw = json.load(f)
    with open(test_path, 'r') as f:
        test_raw = json.load(f)

    def clean(data):
        cleaned = []
        for row in data:
            text = row.get("content") or row.get("text") or row.get("prompt") or row.get("text_input")
            label = int(row.get("label", 0))  # default to 0 if label missing
            if text is not None:
                cleaned.append({"content": text.strip(), "label": label})
        return cleaned

    cleaned_train = clean(train_raw)
    cleaned_test = clean(test_raw)

    # 80/20 split
    train_split, val_split = train_test_split(cleaned_train, test_size=0.2, random_state=42)

    # Save
    with open("train_set_final.json", "w") as f:
        json.dump(train_split, f, indent=2)
    with open("val_set_final.json", "w") as f:
        json.dump(val_split, f, indent=2)
    with open("test_set_final.json", "w") as f:
        json.dump(cleaned_test, f, indent=2)

    print(f"✅ Saved: {len(train_split)} train, {len(val_split)} val, {len(cleaned_test)} test samples")

# Run it
clean_and_split("/content/drive/MyDrive/fp/DeTeCtive/detective_train.json", "/content/drive/MyDrive/fp/DeTeCtive/detective_test.json")
```

```python
import json
import random
import math
import pandas as pd

def convert_list_format_to_dict(path):
    # Load raw data
    raw_data = pd.read_json(path)

    cleaned_data = []
    for row in raw_data.values:
        try:
            text = row[0]
            label = int(row[1])
            source = int(row[3]) if not pd.isna(row[3]) else 0
            if text:
                cleaned_data.append({
                    "content": text,
                    "label": label,
                    "source": source
                })
        except Exception as e:
            print(f"Skipping row due to error: {e}\nRow: {row}")

    return cleaned_data

def split_train_val(data, val_ratio=0.2):
    random.shuffle(data)
    val_size = math.floor(len(data) * val_ratio)
    return data[val_size:], data[:val_size]

# Paths
train_path = "/content/drive/MyDrive/fp/DeTeCtive/detective_train.json"
test_path = "/content/drive/MyDrive/fp/DeTeCtive/detective_test.json"

# Convert and split
train_data_cleaned = convert_list_format_to_dict(train_path)
test_data_cleaned = convert_list_format_to_dict(test_path)

train_split, val_split = split_train_val(train_data_cleaned)

# Save
with open("/content/drive/MyDrive/fp/DeTeCtive/train_set_final.json", "w") as f:
    json.dump(train_split, f, indent=2)

with open("/content/drive/MyDrive/fp/DeTeCtive/val_set_final.json", "w") as f:
    json.dump(val_split, f, indent=2)

with open("/content/drive/MyDrive/fp/DeTeCtive/test_set_final.json", "w") as f:
    json.dump(test_data_cleaned, f, indent=2)

print("✅ Done! Files saved:")
print("- train_set_final.json")
print("- val_set_final.json")
print("- test_set_final.json")
```
