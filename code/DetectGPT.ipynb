# create data

```python
data_path = "/content/drive/MyDrive/fp/train_set_final.parquet"
```

```python
test_data_path = "/content/drive/MyDrive/fp/test_set_final.parquet"
```

```python
import pandas as pd

train_df = pd.read_parquet(data_path)
train_df['label'] = train_df['source']
train_df = train_df[['text','label']]
train_df['label'] = train_df['label'].astype(int)
train_df = train_df.sample(n = 5000,random_state = 42)

train_df
```

```python
import pandas as pd

test_df = pd.read_parquet(test_data_path)
test_df['label'] = test_df['source']
test_df = test_df[['text','label']]
test_df['label'] = test_df['label'].astype(int)
test_df = test_df.sample(n = 2500, random_state=42)

test_df
```

```python
import pandas as pd
import json

# Convert to list of dictionaries
data = [{"text": row["text"], "label": row["label"]} for _, row in test_df.iterrows()]

# Save as JSON
with open("/content/drive/MyDrive/fp/DetectGPT/detect-gpt/test_final.json", "w") as f:
    json.dump(data, f, indent=2)
```
